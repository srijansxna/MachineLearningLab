```python
import pandas as pd
import numpy as np

df = pd.read_csv("StudentsPerformance.csv")


df.head()

numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(exclude=np.number).columns

numerical_cols, categorical_cols

mean_values = df[numerical_cols].mean()
median_values = df[numerical_cols].median()
mode_values = df[numerical_cols].mode().iloc[0]

mean_values, median_values, mode_values

min_values = df[numerical_cols].min()
max_values = df[numerical_cols].max()
sum_values = df[numerical_cols].sum()
variance_values = df[numerical_cols].var()
std_dev_values = df[numerical_cols].std()

min_values, max_values, sum_values, variance_values, std_dev_values
```

```python
Q1 = df[numerical_cols].quantile(0.25)
Q2 = df[numerical_cols].quantile(0.50)  # Median
Q3 = df[numerical_cols].quantile(0.75)

Q1, Q2, Q3
```

```python
correlation_matrix = df[numerical_cols].corr()
correlation_matrix
covariance_matrix = df[numerical_cols].cov()
covariance_matrix
df["math score"].corr(df["reading score"])
df["math score"].cov(df["reading score"])
```
